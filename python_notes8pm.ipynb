# 6.If-Else

```python
#Flow Control stmts
#To execute or not to execute set of stmts
#Flow Control stmts divided into 2 parts
#1.condition :expression which returns a boolean value
#2.Block/Clause :set of stmts following the indentation

#if-else
#syntax for defining if-else
"""
if(condition):
    stmt1
    stmt2------>Block1
    stmt3
else:
    stmt4
    stmt5------>Block2
    stmt6
If the condition is True, then the stmts within the if-block(Block1) executes
If the condition is False, then the stmts within the else-block(Block2) executes
"""
```

```python
#program illustrating if-else
x=4
if(x<=4):
   print("ONE")
   print("TWO")
   print("THREE")
   print("FOUR")
else:
   print("FIVE")
   print("SIX")
   print("SEVEN")
   print("EIGHT")
```

# 8.While loop

```python
#While loop executes set of stmts for repeated number of times or for multiple times
#Syntax for defining while loop:
"""
initialization
while(condition):
   stmt1
   stmt2
   stmt3
   stmt4
   increment/decrement
print("end")
"""
#Stmts within the while loop will keeep on executing untill the condition becomes false
```

```python
#Printing numbers from 1 to 10 using while loop
x=1           #initialization
while(x<=10): #Condition
   print(x)
   x=x+1     #increment
print("end")
```

```python
#while and break stmt
#break stmt: If break stmt is used in a loop then ctrl comes out of that loop and
#stmts after break wont be executed
x=1
while(x<=10):
   print("Hello")
   if(x==5):
      break
      print("Hi")
   x=x+1
print("end")
```

# 9.Lists

```python
#List is a collection datatype holding group of values
#List Properties:
"""
1.List is represented using []

2.List supports heterogeneous elements  ex: x=[101,"Ajay",90870.40]

3.List allows duplicate elements  ex: x=[10,20,30,40,50,10,20,30]

4.List follows the insertion order,i.e the way they are inserted in the same order
  They are stored or displayed
  x=[10,20,30,40,50]

5.List supports 2 types of indexing
  1.Positive indexing(or) Forward indexing-->starts with 0
  2.Negative indexing(or) Backward indexing->starts with -1

     0   1   2   3   4 ------->Forward index/positive index
  x=[10, 20, 30, 40, 50]
     -5  -4  -3  -2  -1 ------>Backward index or negative index

6.List is a mutable object i.e changes or modifications are allowed within a list
```

```python
#Creating a List
x=[10,20,30,40,50,60]
print(x)
print(type(x))
print(len(x))
print(x[0])
print(x[-1])
print(x[-3])
#List slicing
#Extract the 1st 3 elements
print(x[0:3]) #always upperbound is excluded
print(x[ :3])
print(x[3:0])#Empty list-->bcoz always slicing in the forward direction
#Extract last 3 elements
print(x[3:6])
print(x[3: ])
print(x[-3: ])
print(x[ : ])
print(x[:])
print(x)
print(x[0: ])
print(x[::-1]) #reverses the list
```

```python
#List functions
x=[30,10,40,20,50]
print(x)
print(len(x))
print(sum(x))
print(max(x))
print(min(x))
avg=sum(x)/len(x)
print("avg=",avg)
print(sorted(x))
print(sorted(x,reverse=True))
```

```python
#Lists within Lists
players=[["Kohli",90],["Rohith",70],["Gill",20],["Surya",60]]
print(players)
print(len(players))

for p in players:
    print(p,type(p))

for p,q in players:
   print("Player:",p)
   print("Runs:",q)
   print("==========================")

#Compute the total score
tot_score=0
for p,q in players:
    tot_score=tot_score+q
print("Total Score:",tot_score)
```

```python
#Set Mathematical methods
x={10,20,30,40,50,60}
y={10,20,30,70,80,90}

#1.union:
z=x.union(y)
print(z,len(z))

#2.intersection:
p=x.intersection(y)
print(p,len(p))

#3.difference:
q=x.difference(y) #removing y elements from x
#or
r=x-y
print(q,len(q))
print(r,len(r))

a=y-x      #removing x elements from y
#or
b=y.difference(x)
print(a,len(a))
print(b,len(b))

#4.symmetric_difference: Excluding the common elements, remaining all elements
p=x.symmetric_difference(y)
print(p,len(p))

```

```python
#Dictionary is a collection of (k,v) pairs

1.Dictionary also represented as  {}
  x={"name":"Ajay","age":25,"city":"Hyd"}
       k      v      k   v    k      v

2.In Dictinary keys cannot be duplicated but values can be duplicated
  ex:  y={"age":60,"wt":60}

3.Dictionary doesnt support indexing

4.Dictionary is un-ordered

5.Dictionary is a mutable object i.e changes and modifications are allowed
```

```python
x={"name":"Ajay","age":25,"city":"Hyd"}
print(x,type(x))
#If we pass the key--->It returns the value
print(x["name"])
print(x["age"])
print(x["city"])
```

```python
#Modifying,adding,printing values of a dictionary
x={"name":"Ajay","age":25,"city":"Hyd","wt":75}
print(x,len(x))
#modifying the wt to 80
x["wt"]=80
print(x)
#Adding a new element i.e new (k,v) pair -->"designation"
x["designation"]="Manager"
print(x,len(x))

#Adding one more element--->marks
x["marks"]=[["Maths",90],["Phy",80],["Chem",50]]
print(x,len(x))

#Modify the marks-->80 to 70
x["marks"][1][1]=70
print(x,len(x))

#Dictionary within a dictionary
y={"Name":"Ajay","age":40,"children":{"ch1":10,"ch2":6},"address":{"city":"Hyd","state":"Telangana"}}
print(y,len(y))
#print--->"Hyd"--->we need to pass 2 keys
print(y["address"]["city"])
```

```python
#Dictionary methods

x={"name":"Ajay","age":25,"city":"Hyd","wt":75}
print(x,len(x))
#1.print all the keys only---->keys()
print(x.keys())
#2.printing only the values--->values()
print(x.values())

#3.pop()Removing an element from a dictionary
x.pop("city")
print(x,len(x))

#4.popitem() :Removes an element randomly
x.popitem()
print(x,len(x))

#5.clear():Clears all the element of a dictionary
x.clear()
print(x,len(x))

#adding a new element
x["name"]="Amar"
print(x,len(x))
```

```python
#Lambda Function:

# A Function without any name is called as Lambda function

# Generally a Function called with a Name

# Synatax for defining lambda function

# lambda arguments:expression

# If a function doesnt have a name then how to call the function?

# Assign the function to a variable and the variable behaves like the function name


#ex:1  Lambda function to square a number
f1=lambda x:x**2  #assign this lambda function to a variable
print(f1(10))
print(f1(20))

#Same Task using Normal function

def f2(x):
    return x**2
print(f2(10))
print(f2(20))
```

```python
#we can also provide multiple arguments to lambda function but only one expression
p=lambda x,y:x*y
print(p(10,20))
```

```python
#Why lambda function??

#If you want to provide a function as a argument to another function
#then go for Lambda function

#we see how lambda function used as an argument to other function such as filter(),map()
```

```python
#filter function :For filtering data based on condition

#filter function takes 2 arguments
#1.lambda function
#2.Iterable type like list

#ex:1
list1=[10,20,30,40,50,60]
#Task: Filter only those elements whose value>30
res=list(filter(lambda x:x>30,list1)) #here x is each element of list of type--->int
print(res,type(res))

#ex:2
emps=[[101,"miller",10000,'m',11,'hyd'],
      [102,"blake",20000,'m',12,'pune'],
      [103,"sony",30000,'f',11,'hyd'],
      [104,"sita",40000,'f',12,'pune'],
      [105,'james',50000,'m',13,'hyd']]

#Task1 : Filter only the emps who belongs to the city "hyd"
res=list(filter(lambda x:x[5]=="hyd",emps)) #here x is each element of list of type list only
print(res,type(res))

#Task 2: Filter only male emp records
res2=list(filter(lambda x:x[3]=='m',emps))
print(res2)
print("\n")

#providing multiple conditions
#Task 3:Filter those emps whose sal>20000 and dno=12
res3=list(filter(lambda x:x[2]>20000 and x[4]==12,emps))
print(res3)
```

```python
#map function :For each element transformation

#map function takes 2 arguments
#1.lambda function
#2.Iterable type like list

#ex:1
list1=[10,20,30,40,50,60]
#Task: adding 5 to each element of the list
res1=list(map(lambda x:x+5,list1))
print(res1,type(res1))

#ex:2
emps=[[101,"miller",10000,'m',11,'hyd'],
      [102,"blake",20000,'m',12,'pune'],
      [103,"sony",30000,'f',11,'hyd'],
      [104,"sita",40000,'f',12,'pune'],
      [105,'james',50000,'m',13,'hyd']]

#Task 1: adding 5000 to the sals of each employee as bonus

res2=list(map(lambda x:[x[0],x[1],x[2]+5000,x[3],x[4],x[5]],emps))
print(res2)
print("\n")

#Task2 :Transforming each empname to start with uppercase character
res2=list(map(lambda x:[x[0],x[1].title(),x[2],x[3],x[4],x[5]],emps))
print(res2)
```

```python
#Various File operations are
#1.opening a file
#2.closing a file
#3.Reading data from a file
#4.writing data to a file
#5.appending data to a file

#For all the above operations,we have in-built methods or functions such as
#open(),close(),read(),write()
#open() :whenever we call open function by providing the filelocation then
#        file object is created
#ex:  f=open("filename",mode="r")  here f is the file object
#whenever we open a file, we need to specify the mode
#mode indicates for what purpose you are opening the file

#Various modes

#1.r  ------->read mode(default mode)

#2.w ----->write mode
#          i)if file already exists-->it removes(truncates) the entire data and writes
#         ii)if file doesnt exists--->it creates new file automatically and writes

#3.a ----->append mode
#         i)if file already exists-->it appends data at the end of the file
#         ii)if file doesnt exists--->it creates new file and appends

#4.w+ ---->writes and reads
#        -here previous data in the file will be removed

#5.r+ ---->reads and writes
#        -here previous data in the file wont be removed

#6.a+ --->appends and reads
```

```python
#Dataframe filtering operations

#Filtering data from a Dataframe

# importing pandas
import pandas as pd

data = {
 'Name': ['Miller','Blake','James','Jimmy','Sony','Ruby' ],
 'Age': [23,24,22,25,23,22],
 'Gender':['m','m','m','f','f','f'],
 'Branch': ['CSE','ECE','EEE','IT','CSE','ECE'],
 'Percentage': [84,82,78,76,68,56],
 }
# create a dataframe
df = pd.DataFrame(data,columns = ['Name', 'Age','Gender','Branch', 'Percentage'])
print(df)
print("\n")

#-------------------------------------------------------------------------------------------------------
#i.filtering data based on percentage

df1 = df[df['Percentage'] > 80]
print(df1)
#Note: If we filter a Dataframe then the result is also a Dataframe
print("\n")

#or

df1 = df.loc[df['Percentage'] > 80]
print(df1)

print("\n")
#------------------------------------------------------------------------------------------------------

#ii. I want only the CSE students

df1 = df[df['Branch'] == 'CSE']
print(df1)

print("\n")
#------------------------------------------------------------------------------------------------------
#iii. I want other than CSE students
df1 = df[df['Branch'] != 'CSE']
print(df1)

print("\n")

#Note : we can use == , != , >= ,<= ,> , < for filtering the data

#-----------------------------------------------------------------------------------------------------
#iv : isin() method()

#ex:
#I require only those students who belongs to CSE or IT background

options=['CSE','IT']

df3 = df[df['Branch'].isin(options)]
print(df3)

print("\n")

#or
df4 = df.loc[df['Branch'].isin(options)]
print(df4)

print("\n")
#------------------------------------------------------------------------------------------------------

#V :I want to know those students who are from non-CSE background
options=['CSE','IT']

df5 = df[~df['Branch'].isin(options)]
print(df5)
print("\n")
#------------------------------------------------------------------------------------------------------

#Vi. Providing multiple conditions
#I want those students whose percentage < 70 and are non-CSE Background
df6 = df[(df['Percentage'] < 70) &
        ~df['Branch'].isin(options)]
print(df6)

```

```python
#Groupings and Aggregations in Pandas
import pandas as pd

df = pd.read_csv('/content/orders.csv') #df is a Dataframe
#print (df)

#Reading Specific Rows of a particular column--->5rows
print (df[0:5]['customer'])

#Reading Specific Columns
'''We use the multi-axes indexing method called loc() for this purpose.
We choose to display the customer and order column for all the rows.'''
#print (df.loc[:,['customer','order']])

#Reading Specific Columns and Rows
'''The read_csv function of the pandas library can also be used to read some specific columns and specific rows.
We use the multi-axes indexing method called .loc() for this purpose.
We choose to display the customer and order column for some of the rows.'''
print(df.loc[[1,3,5],['customer','order']])

#Reading Specific Columns for a Range of Rows
print (df.loc[2:6,['customer','order']])

#------------------------------------------------------------------------------

#Various Aggregations--->sum(),mean(),max(),min(),count()

#How many rows the dataset has i.e how many total orders were made--->using count()
#1.No of orders----->using count()
orders=df['order'].count()
print("NO OF ORDERS:",orders)

#2.Total order value(TOV) generated for all the orders ---->sum()
tot_rev=df['revenue'].sum()
print("TOTAL ORDER VALUE:",tot_rev)

#3.Average order value(AOV) generated for all orders ---->mean()
avg_rev=df['revenue'].mean()
print("AVERAGE ORDER VALUE(AOV):",avg_rev)

#4.Max order value  ---->max()
max_rev=df['revenue'].max()
print("MAX ORDER VALUE:",max_rev)

#5.Min order value  ---->min()
min_rev=df['revenue'].min()
print("MIN ORDER VALUE:",min_rev)

# TASK-1. Calculate total number of orders per customer;
orders_cust=df.groupby('customer')['order'].count()
#same thing in sql--->
#select customer,count(order) from orders group by customer
print("Total No of orders per customer:\n",orders_cust)

#TASK-2: Calculate TOV per customer, TOV  - Total revenue;
TOV_cust=df.groupby('customer')['revenue'].sum()
print("TOV per Customer:\n",TOV_cust)

#TASK-3: Calculate AOV per customer, AOV  - average revenue;
AOV_cust=df.groupby('customer')['revenue'].mean()
print("AOV per Customer: \n",AOV_cust)

# Calculate the count of orders based on order_state
cnt=df.groupby('order_state')['order_state'].count()
print("order_state:\n",cnt)



#multi grouping
#df.groupby(['col1', 'col2']).agg({'col3': [min, max, sum]})
#o/p:
'''
citywise,dnowise---->totsal
hyd 11 --->totsal
hyd 12 --->totsal
hyd 13 --->totsal
pune 11 -->totalsal
pune 12 -->totsal
pune 13 -->totsal
'''
df2 = pd.read_csv('emp1.csv')
df3=df2.groupby(['city','dno']).agg({'Salary':[sum]})
print(df3)
#o/p:
'''
city dno
city dno
hyd  11   10000
     12   30000
     13   50000
pune 11   40000
     12   20000
'''
#-------------------------------------------------------------------------------

#Filtering data:
#Selecting rows in pandas DataFrame based on conditions

#I want only those orders whose ordervalue(revenue) >100
df1 = df[df['revenue'] > 100]
print(df1)
print("\n")
#or
df2 = df.loc[df['revenue'] > 100]
print(df2)
print("\n\n\n")

print("\n")

#I want those orders which are not completed
#ex:2
df3 = df.loc[df['order_state']!='completed']
print(df3)
```
