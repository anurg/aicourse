```python
class Dog():
    '''  This class represents Dog and its .
    '''
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def print_dog(self):
        print(f"My Dog name is :{self.name} and age:{self.age}")
        
    def sit(self):
        '''  Simulates the dog sitting           '''
        print(self.name.title(), "is now sitting.")
    def roll_over(self):
        ''' Simulates the dog rolling over  '''
        print(self.name.title(), "is rolling over.")
    def update_age(self,years_passed):
        ''' Increments the age by No of years Passed '''
        self.age += years_passed

shadow = Dog("Shadow",12)
# print(f"My Dog name is :{shadow.name} and age:{shadow.age}")
shadow.print_dog()
shadow.sit()
shadow.roll_over()
print("One year later")
shadow.update_age(1)
print(f"My Dog name is :{shadow.name} and age:{shadow.age}")
```

```python
class Labrador(Dog):
    def __init__(self,name,age,color):
        ''' Special Dog breed Labrador '''
        super().__init__(name,age)
        self.color = color
    def print_dog(self):
        print(f"My Dog name is :{self.name} and age:{self.age} and color: {self.color}")
my_dog = Labrador("Boozo",10,"Brown")
my_dog.print_dog()
        
```
