```python
alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}
aliens = [alien_0, alien_1, alien_2]
for alien in aliens:
    print(alien)
```

```python
aliens =[]
for i in range(1,10):
    new_alien= {"id":i,"color":"green","points":5,"speed":"slow"}
    aliens.append(new_alien)
for alien in aliens:
    print(alien)
```
