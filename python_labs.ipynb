```python
#Lists within Lists
players=[["Kohli",90],["Rohith",70],["Gill",20],["Surya",60]]
print(players,type(players))
print(players[0][0])
print(players[0][1])
print(len(players))
for p in players:
  print(p,type(p))
total_score =0
for p,q in players:
  # print("Player Name: ",p," -- Player Avg: ",q)
  print("Player Name: ",p)
  print("Player Avg: ",q)
  total_score +=q
  print("===========================")
print("Total Score: ",total_score)
```
