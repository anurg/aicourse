```python
x = [1,2,3,4,5]
print(x,id(x),type(x))
x[2] +=3
print(x,id(x),type(x))
t = (1,2,3)
print(t,id(t),type(t))
print(t[2])
s={1,2,2,3,4,5}
print(s,id(s),type(s))
d={'a':1,'b':2,'c':3}
print(d,id(d),type(d))
```

```python
#Strings
s= "Hello Python!"
print(s,type(s))
print(s[2],type(s[2]))
print(s[-2],type(s[-2]))
```

```python
### String methods
print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
print(s.count("l"))
print(s.replace("Hello","Hi"))
x = "{} is the PM of {}"
print(x.format("Modi","India"))
print(x.format("Maccron","France"))
print(x.find("M"))
print(x.find("Z"))
y = x.split(" ")
print(y,type(y))
z = " Good Evening!    "
print(z.strip())
```
