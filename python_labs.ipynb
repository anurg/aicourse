```python
#### if elif else
import time
print(time.time())
print(time.localtime())
t = float(input("Please enter time: "))
if t<=12.00:
  print("Good Morning")
elif t<=16.00:
  print("Good Afternoon")
elif t<=20.00:
  print("Good Evening")
else:
  print("Good Night")
```
