```python
# matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10,20,30,40,50])
y=np.array([23,45,34,56,34])
plt.plot(x,y,"--",color="green")
plt.axis([0,70,0,60]) #[xmin,xmax,ymin,ymax]
                     #[xmin to xmax,ymin to ymax]
plt.title("First Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#draw a line between 2 points (2,7) and (3,8)
plt.plot([2,3],[7,8])
plt.show()
```

```python
#Horizontal Line & Vertical line
x=[10,20,30,40,50]
y=[30,30,30,30,30]
plt.plot(x,y)
plt.plot(y,x)
plt.show()
```
