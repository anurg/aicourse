```python
# matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10,20,30,40,50])
y=np.array([23,45,34,56,34])
plt.plot(x,y,"--",color="green")
plt.axis([0,70,0,60]) #[xmin,xmax,ymin,ymax]
                     #[xmin to xmax,ymin to ymax]
plt.title("First Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#draw a line between 2 points (2,7) and (3,8)
plt.plot([2,3],[7,8])
plt.show()
```

```python
#Horizontal Line & Vertical line
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[30,30,30,30,30]
plt.plot(x,y,linestyle="dashed",linewidth=4,label="Vertical Line",color="brown")
plt.plot(y,x,linestyle="dotted",linewidth=4,label="Vertical Line",color="green")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
plt.plot(x,marker="o",label="trend")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("First \n Plot")
# plt.grid()
# plt.grid(axis="x")
plt.grid(axis="y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[1,2,3]
y1=[5,7,4]

x2=[1,2,3]
y2=[10,14,8]
plt.plot(x1,y1,label="line1")
plt.plot(x2,y2,label="line2")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
overs=[10,20,30,40,50]
runs_india=[75,110,195,230,290]
runs_pak =[50,120,155,205,260]
plt.plot(overs,runs_india,label="India",marker="o",color="blue")
plt.plot(overs,runs_pak,"--",label="Pakistan",color="green")
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.title("Runs Comparison")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.barh([1,3,5,7,9],[9,3,7,5,6],label="LG")
plt.barh([2,4,6,8,10],[6,9,5,1,2],label="Sony",color="g")
plt.xlabel("Months")
plt.ylabel("Sales in Units")
plt.title("Sales Comparison")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x= ["Team-A","Team-B","Team-C","Team-D"]
y1=[10,20,10,35]
y2=[10,30,15,20]
plt.bar(x,y1,label="Round1",color="b")
plt.bar(x,y2,bottom=y1,label="Round2",color="g")
plt.show()
```

```python
import matplotlib.pyplot as plt

player_score=[30,40,10,20]
players=["Rohit","Virat","Dhoni","Hardik"]
plt.pie(player_score,labels=players,startangle=90,explode=(0,0.2,0,0),shadow=True,autopct='%1.2f%%')
plt.title("Team Runs")
plt.legend()
plt.show()
```
