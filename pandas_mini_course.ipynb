# Module 1: The DataFrame Mental Model
*Estimated time: 45 minutes*

## 🧠 CONCEPT: DataFrame as a labeled 2D structure

A DataFrame is a table with labeled rows and labeled columns. Under the hood, each column is a NumPy array (wrapped in a Pandas Series). The labels are what make Pandas powerful — you access data by meaningful names instead of remembering position numbers.

## 🧠 CONCEPT: Series vs DataFrame

```python
import pandas as pd

# Series = one column with an index (labels)
prices = pd.Series([100, 105, 102], index=['Mon', 'Tue', 'Wed'])
print("Series:")
print(prices)
print(f"\nAccess by label: prices['Tue'] = {prices['Tue']}")

# DataFrame = multiple Series sharing the same index
df = pd.DataFrame({
    'open':   [100, 105, 102],
    'close':  [105, 102, 108],
    'volume': [1000, 1500, 800]
}, index=['Mon', 'Tue', 'Wed'])
print("\nDataFrame:")
print(df)
```

### Critical insight: Single column → Series, Multiple columns → DataFrame

```python
import pandas as pd

df = pd.DataFrame({
    'open':   [100, 105, 102],
    'close':  [105, 102, 108],
    'volume': [1000, 1500, 800]
}, index=['Mon', 'Tue', 'Wed'])

print("Single column (Series):")
print(type(df['open']))
print(df['open'])

print("\nMultiple columns (DataFrame):")
print(type(df[['open', 'close']]))  # Note the double brackets!
print(df[['open', 'close']])
```

# Module 2: Selection and Filtering — Getting the Data You Need
*Estimated time: 60 minutes*

## 🧠 CONCEPT: loc vs iloc vs []

- **`df[...]` (bracket notation):** For selecting columns by name or filtering rows with boolean conditions
- **`df.loc[row_label, col_label]`:** Label-based. INCLUSIVE on both ends.
- **`df.iloc[row_position, col_position]`:** Position-based (like NumPy). EXCLUSIVE on the end.

```python
import pandas as pd

# Sample data
df = pd.DataFrame({
    'name':   ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept':   ['Eng', 'Mktg', 'Eng', 'Mktg'],
    'salary': [90000, 80000, 95000, 85000]
}, index=['emp1', 'emp2', 'emp3', 'emp4'])

print("DataFrame:")
print(df)

# Bracket notation
print("\n--- Bracket Notation ---")
print("df['salary']:\n", df['salary'])                    # One column (Series)
print("\ndf[['name', 'salary']]:\n", df[['name', 'salary']])  # Multiple columns (DataFrame)
print("\ndf[df['salary'] > 85000]:\n", df[df['salary'] > 85000])  # Boolean filter

# .loc — by labels
print("\n--- .loc (Label-Based) ---")
print("df.loc['emp2']:\n", df.loc['emp2'])                # Row by label
print("\ndf.loc['emp1':'emp3', 'name']:\n", df.loc['emp1':'emp3', 'name'])  # INCLUSIVE!
print("\ndf.loc[df['dept'] == 'Eng']:\n", df.loc[df['dept'] == 'Eng'])

# .iloc — by position
print("\n--- .iloc (Position-Based) ---")
print("df.iloc[0]:\n", df.iloc[0])                        # First row
print("\ndf.iloc[1:3, 0:2]:\n", df.iloc[1:3, 0:2])       # Rows 1-2, Cols 0-1 (EXCLUSIVE end)
```

## 🧠 CONCEPT: Boolean Filtering — Your SQL WHERE Clause

This is the Pandas operation you'll use most.

```python
import pandas as pd

df = pd.DataFrame({
    'name':   ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'dept':   ['Eng', 'Mktg', 'Eng', 'Mktg', 'Sales'],
    'salary': [90000, 80000, 95000, 85000, 72000]
})

# Single condition
print("High earners (salary > 85000):")
print(df[df['salary'] > 85000])

# Multiple conditions (& for AND, | for OR, ~ for NOT)
# IMPORTANT: Parentheses around each condition are MANDATORY
print("\nEngineers with high salary:")
print(df[(df['dept'] == 'Eng') & (df['salary'] > 85000)])

# Using .isin() for multiple matches
print("\nEng or Sales departments:")
print(df[df['dept'].isin(['Eng', 'Sales'])])

# String methods
print("\nNames starting with 'A':")
print(df[df['name'].str.startswith('A')])

print("\nNames containing 'li' (case-insensitive):")
print(df[df['name'].str.contains('li', case=False)])
```

### Smell Test #2

AI generates `df[df.salary > 85000 & df.dept == 'Eng']`. What's wrong?

**Missing parentheses!** Python evaluates `&` before `>` and `==`, causing an error. Correct: `df[(df['salary'] > 85000) & (df['dept'] == 'Eng')]`

# Module 3: GroupBy — Split-Apply-Combine
*Estimated time: 60 minutes*

## 🧠 CONCEPT: Split-Apply-Combine pattern

GroupBy is Pandas' most powerful analytical tool:
1. **Split:** Divide data into groups based on column(s)
2. **Apply:** Run a function on each group independently
3. **Combine:** Merge the results back into a single structure

```python
import pandas as pd

# Sales data
df = pd.DataFrame({
    'region':  ['North', 'South', 'North', 'South', 'North'],
    'product': ['A', 'A', 'B', 'B', 'A'],
    'revenue': [100, 150, 200, 120, 180],
    'cost':    [60, 90, 110, 80, 100]
})
print("Data:")
print(df)

# Basic groupby
print("\n--- Mean revenue by region ---")
print(df.groupby('region')['revenue'].mean())

# Multiple aggregations at once
print("\n--- Multiple aggregations ---")
print(df.groupby('region')['revenue'].agg(['mean', 'sum', 'count']))

# Different aggregations per column
print("\n--- Different aggs per column ---")
print(df.groupby('region').agg({
    'revenue': 'sum',
    'cost': 'mean'
}))

# Multiple grouping columns
print("\n--- Group by region AND product ---")
print(df.groupby(['region', 'product'])['revenue'].sum())
```

## 🧠 CONCEPT: Transform vs Aggregate

**Aggregate** reduces each group to a single value (like SUM, MEAN).
**Transform** returns a value for EVERY ROW (same shape as input).

```python
import pandas as pd

df = pd.DataFrame({
    'region':  ['North', 'South', 'North', 'South', 'North'],
    'product': ['A', 'A', 'B', 'B', 'A'],
    'revenue': [100, 150, 200, 120, 180],
    'cost':    [60, 90, 110, 80, 100]
})

# Aggregate: one row per group
print("Aggregate (sum per region):")
print(df.groupby('region')['revenue'].sum())

# Transform: value for every row (broadcasts back)
df['region_avg'] = df.groupby('region')['revenue'].transform('mean')
print("\nTransform (region avg added to each row):")
print(df[['region', 'revenue', 'region_avg']])

# Common pattern: "what % of my region's total am I?"
df['pct_of_region'] = df['revenue'] / df.groupby('region')['revenue'].transform('sum')
print("\nPercentage of region total:")
print(df[['region', 'revenue', 'pct_of_region']].round(3))
```

# Module 4: Handling Messy Data — The Reality of Data Science
*Estimated time: 60 minutes*

## 🧠 CONCEPT: Data cleaning is 80% of the work

This isn't an exaggeration. Real datasets have missing values, wrong types, duplicates, inconsistent formatting, and outliers.

## 🧠 CONCEPT: Strategies for missing values

Three approaches:
- **Drop it:** Remove rows/columns with missing data
- **Fill it:** Replace with something (mean, median, mode, forward fill)
- **Flag it:** Create a separate column indicating "this was missing"

```python
import pandas as pd
import numpy as np

# Create messy dataset with missing values
rng = np.random.default_rng(42)
df = pd.DataFrame({
    'employee':  ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'revenue':   [50000, np.nan, 75000, 60000, np.nan, 80000],
    'dept':      ['Eng', 'Mktg', np.nan, 'Eng', 'Sales', 'Mktg'],
    'rating':    [4.5, 3.8, 4.2, np.nan, 3.5, np.nan]
})

print("Original messy data:")
print(df)

# Detecting missing data
print("\n--- Missing Value Counts ---")
print(df.isnull().sum())
print("\n--- Missing Value Percentages ---")
print(df.isnull().mean().round(3))

# Dropping
print("\n--- dropna(): Drop rows with ANY NaN ---")
print(df.dropna())

print("\n--- dropna(subset=['revenue']): Only if revenue is NaN ---")
print(df.dropna(subset=['revenue']))

# Filling
print("\n--- fillna with median (revenue) ---")
df_filled = df.copy()
df_filled['revenue'] = df_filled['revenue'].fillna(df_filled['revenue'].median())
print(df_filled[['employee', 'revenue']])

# Flagging
print("\n--- Flag missing as separate column ---")
df['revenue_missing'] = df['revenue'].isnull().astype(int)
print(df[['employee', 'revenue', 'revenue_missing']])
```

## 🧠 CONCEPT: Data Types — The Silent Bug Factory

When Pandas reads a CSV, it guesses column types. Numbers with commas become strings. A single "N/A" makes a whole column string.

```python
import pandas as pd
import numpy as np

# Simulate messy CSV data with type problems
df = pd.DataFrame({
    'price':    ['100.50', '200.75', 'N/A', '150.25', '180.00'],
    'date':     ['2024-01-15', '2024-02-20', '2024-03-10', '2024-04-05', '2024-05-01'],
    'category': ['A', 'B', 'A', 'C', 'B'],
    'quantity': ['10', '20', '15', '8', '25']
})

print("Before fixing types:")
print(df.dtypes)
print()

# Fix types
df['price'] = pd.to_numeric(df['price'], errors='coerce')  # Invalid → NaN
df['date'] = pd.to_datetime(df['date'])
df['category'] = df['category'].astype('category')  # Memory efficient
df['quantity'] = pd.to_numeric(df['quantity'])

print("After fixing types:")
print(df.dtypes)
print()
print(df)
```

## 👀 RECOGNIZE: Duplicates and Inconsistencies

```python
import pandas as pd

# Messy data with duplicates and inconsistent strings
df = pd.DataFrame({
    'id':   [1, 2, 3, 2, 4, 5],
    'city': ['  New York', 'new york', 'LONDON', 'New York  ', 'london', 'Paris'],
    'state': ['CA', 'Cal', 'NA', 'California', 'NA', 'NA']
})

print("Original messy data:")
print(df)

# Duplicates
print(f"\nDuplicate rows (by id): {df.duplicated(subset=['id']).sum()}")
df_deduped = df.drop_duplicates(subset=['id'])
print(f"After dedup: {len(df_deduped)} rows")

# String inconsistencies — normalize
df['city'] = df['city'].str.strip().str.lower()
df['state'] = df['state'].replace({'CA': 'California', 'Cal': 'California'})
print("\nAfter cleaning strings:")
print(df)
```

### Smell Test #3

# Module 5: Merging and Joining — Combining Data Sources
*Estimated time: 60 minutes*

## 🧠 CONCEPT: Relational thinking

Real analysis almost always requires combining data from multiple sources.

## 🧠 CONCEPT: The Four Join Types

| Join Type | What It Keeps | When To Use |
|---|---|---|
| **inner** | Only rows matching in BOTH tables | When you only want complete data |
| **left** | ALL from left, matched right where possible | Preserve all records from main table |
| **right** | ALL from right, matched left where possible | Rare — swap order and use left |
| **outer** | ALL rows from BOTH tables | When you can't afford to lose any data |

```python
import pandas as pd

# Two DataFrames
customers = pd.DataFrame({
    'cust_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana']
})

orders = pd.DataFrame({
    'cust_id': [2, 3, 3, 5],
    'amount': [100, 200, 150, 300]
})

print("Customers:")
print(customers)
print("\nOrders:")
print(orders)

# Inner join: only customers WITH orders
print("\n--- Inner Join ---")
print(pd.merge(customers, orders, on='cust_id', how='inner'))

# Left join: ALL customers, orders where available (NaN otherwise)
print("\n--- Left Join ---")
print(pd.merge(customers, orders, on='cust_id', how='left'))

# Outer join: everyone
print("\n--- Outer Join ---")
print(pd.merge(customers, orders, on='cust_id', how='outer'))
```

## 👀 RECOGNIZE: Concat vs Merge

**Concat** stacks DataFrames (same columns, more rows — like SQL UNION).
**Merge** relates DataFrames (shared key, combine columns — like SQL JOIN).

```python
import pandas as pd

# Concat: stacking same-structure data
jan = pd.DataFrame({'date': ['Jan 1', 'Jan 2'], 'sales': [100, 110]})
feb = pd.DataFrame({'date': ['Feb 1', 'Feb 2'], 'sales': [150, 130]})

all_months = pd.concat([jan, feb], ignore_index=True)
print("Concat (stacking):")
print(all_months)
```

### Smell Test #4

You merge two tables and your row count increases unexpectedly — say from 1000 to 3500. **What happened?** Duplicate keys causing a many-to-many join. Each match multiplies rows. **Always check `df.shape` after a merge.**

```python
import pandas as pd

# Demonstrate the row explosion problem
products = pd.DataFrame({
    'category': ['Electronics', 'Electronics', 'Books'],  # Duplicate key!
    'product': ['Phone', 'Laptop', 'Novel']
})
prices = pd.DataFrame({
    'category': ['Electronics', 'Electronics', 'Books'],  # Duplicate key!
    'price': [999, 1499, 15]
})

merged = pd.merge(products, prices, on='category')
print(f"Products: {len(products)} rows")
print(f"Prices:   {len(prices)} rows")
print(f"Merged:   {len(merged)} rows  ← ROW EXPLOSION!")
print()
print(merged)
```

# Module 6: Reshaping — Pivot, Melt, and Pivot Tables
*Estimated time: 45 minutes*

## 🧠 CONCEPT: Long vs Wide format

- **Long format:** date, stock, price — each row is one observation
- **Wide format:** date, AAPL, GOOG, MSFT — each column is a stock

```python
import pandas as pd

# Long to Wide: pivot
long_df = pd.DataFrame({
    'date':  ['Jan', 'Jan', 'Feb', 'Feb'],
    'stock': ['AAPL', 'GOOG', 'AAPL', 'GOOG'],
    'price': [150, 2800, 155, 2750]
})
print("Long format:")
print(long_df)

wide_df = long_df.pivot(index='date', columns='stock', values='price')
print("\nWide format (after pivot):")
print(wide_df)

# Wide to Long: melt
long_again = wide_df.reset_index().melt(
    id_vars='date', var_name='stock', value_name='price'
)
print("\nBack to Long (after melt):")
print(long_again)
```

### Pivot Table: pivot + aggregation (like Excel pivot tables)

Key distinction: `pivot` requires unique index-column combinations. `pivot_table` handles duplicates by aggregating.

```python
import pandas as pd

# Data with duplicate combinations (pivot would fail, pivot_table works)
df = pd.DataFrame({
    'region':  ['North', 'South', 'North', 'South', 'North', 'South'],
    'product': ['A', 'A', 'B', 'B', 'A', 'A'],
    'revenue': [100, 150, 200, 120, 180, 90]
})

print("Data (note: duplicate region-product combos):")
print(df)

print("\nPivot Table (aggregates duplicates with sum):")
print(df.pivot_table(
    values='revenue',
    index='region',
    columns='product',
    aggfunc='sum',
    fill_value=0
))
```

# Module 7: Time Series — Working With Dates
*Estimated time: 45 minutes*

## 🧠 CONCEPT: Time as a special dimension

Pandas has exceptional time series support. This connects directly to your investing book work.

## 🧠 CONCEPT: DatetimeIndex — The Key

```python
import pandas as pd
import numpy as np

# Generate dummy stock data (simulating what you'd read from CSV)
rng = np.random.default_rng(42)
dates = pd.date_range('2023-01-02', periods=500, freq='B')  # Business days
prices = 100 + np.cumsum(rng.normal(0.05, 1.5, 500))

df = pd.DataFrame({'price': prices}, index=dates)
df.index.name = 'date'

print("Stock data with DatetimeIndex:")
print(df.head())
print(f"\nShape: {df.shape}")
print(f"Index type: {type(df.index)}")

# Now you can slice by date strings!
print("\n--- All of 2024 ---")
print(df.loc['2024'].head())

print("\n--- March to June 2024 ---")
print(df.loc['2024-03':'2024-06'].head())

print("\n--- Specific date ---")
print(df.loc['2024-01-15':'2024-01-16'])
```

## 🧠 CONCEPT: Resampling — Changing Time Granularity

Resampling is like `groupby` for time. **Downsampling** = finer to coarser (daily → monthly).

```python
import pandas as pd
import numpy as np

# Generate stock data
rng = np.random.default_rng(42)
dates = pd.date_range('2023-01-02', periods=500, freq='B')
prices = 100 + np.cumsum(rng.normal(0.05, 1.5, 500))
df = pd.DataFrame({'price': prices}, index=dates)

# Daily → Monthly (downsampling)
monthly = df['price'].resample('ME').mean()
print("Monthly average prices:")
print(monthly.head())

monthly_ohlc = df['price'].resample('ME').ohlc()
print("\nMonthly OHLC:")
print(monthly_ohlc.head())

# Rolling windows (moving averages)
df['MA_20'] = df['price'].rolling(window=20).mean()
df['MA_50'] = df['price'].rolling(window=50).mean()

# Pct change (daily returns)
df['returns'] = df['price'].pct_change()

# Cumulative returns
df['cumulative'] = (1 + df['returns']).cumprod()

print("\nDataFrame with technical indicators:")
print(df.tail(10).round(4))

# For your investing work: rolling means and pct_change are the workhorses.
# A 20-day vs 50-day MA crossover is a classic trading signal!
```

# Module 8: Method Chaining — Writing Clean Pandas
*Estimated time: 30 minutes*

## 🧠 CONCEPT: Readable data pipelines

Method chaining makes your code read like a recipe — which is exactly how you should think about data transformations.

```python
import pandas as pd
import numpy as np

# Create sample data
rng = np.random.default_rng(42)
df = pd.DataFrame({
    'region':  rng.choice(['North', 'South', 'East', 'West'], 200),
    'product': rng.choice(['Widget', 'Gadget', 'Doohickey'], 200),
    'revenue': rng.normal(1000, 300, 200).round(2),
    'cost':    rng.normal(600, 200, 200).round(2)
})
# Inject some NaN
df.loc[rng.choice(200, 10, replace=False), 'revenue'] = np.nan

print("Sample data:")
print(df.head())
print(f"Shape: {df.shape}, NaN in revenue: {df['revenue'].isna().sum()}")

# WITHOUT chaining (hard to follow):
df2 = df.dropna(subset=['revenue'])
df3 = df2[df2['region'] == 'North']
df4 = df3.groupby('product')['revenue'].sum()
df5 = df4.sort_values(ascending=False)
result_old = df5.reset_index()
print("\nOld style result:")
print(result_old)

# WITH chaining (reads like English):
result_chain = (
    df
    .dropna(subset=['revenue'])
    .query("region == 'North'")
    .groupby('product')['revenue']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
print("\nChained result (same output, cleaner code):")
print(result_chain)
```

# Capstone Project: Investment Portfolio Dashboard Data
*Estimated time: 3-4 hours*

This connects directly to your "Investing for Programmers" book and uses every module.

**Requirements:**
1. Data Loading & Exploration (Module 1)
2. Data Cleaning (Module 4)
3. Feature Engineering — returns, MAs, volatility, Bollinger Bands (Modules 7, 2)
4. Filtering — find golden cross signals (Module 2)
5. Multi-Stock Comparison (Modules 5, 6)
6. Sector Analysis (Module 3)
7. Method Chaining throughout (Module 8)

```python
import pandas as pd
import numpy as np

# === 1. Generate multi-stock data (Module 1) ===
rng = np.random.default_rng(42)
dates = pd.date_range('2023-01-02', periods=500, freq='B')

stocks = {
    'AAPL':  {'sector': 'Tech',       'start': 150, 'drift': 0.05, 'vol': 1.8},
    'GOOG':  {'sector': 'Tech',       'start': 130, 'drift': 0.04, 'vol': 2.0},
    'JPM':   {'sector': 'Finance',    'start': 140, 'drift': 0.03, 'vol': 1.5},
    'JNJ':   {'sector': 'Healthcare', 'start': 160, 'drift': 0.02, 'vol': 1.0},
    'XOM':   {'sector': 'Energy',     'start': 100, 'drift': 0.01, 'vol': 2.2},
}

# Build long-format DataFrame
records = []
for ticker, params in stocks.items():
    prices = params['start'] + np.cumsum(rng.normal(params['drift'], params['vol'], 500))
    for i, d in enumerate(dates):
        records.append({'date': d, 'ticker': ticker, 'sector': params['sector'], 'price': prices[i]})

df = pd.DataFrame(records)
print("=== Exploration (Module 1) ===")
print(df.head(10))
print(f"Shape: {df.shape}")
print(f"Tickers: {df['ticker'].unique()}")

# === 2. Feature Engineering (Module 7) ===
# Add returns, MAs per stock using transform
df = df.sort_values(['ticker', 'date'])
df['returns'] = df.groupby('ticker')['price'].pct_change()
df['MA_20'] = df.groupby('ticker')['price'].transform(lambda x: x.rolling(20).mean())
df['MA_50'] = df.groupby('ticker')['price'].transform(lambda x: x.rolling(50).mean())

# === 3. Find Golden Cross signals (Module 2) ===
df['prev_MA_20'] = df.groupby('ticker')['MA_20'].shift(1)
df['prev_MA_50'] = df.groupby('ticker')['MA_50'].shift(1)
golden_cross = df[
    (df['MA_20'] > df['MA_50']) &
    (df['prev_MA_20'] <= df['prev_MA_50'])
][['date', 'ticker', 'price', 'MA_20', 'MA_50']]
print("\n=== Golden Cross Signals (Module 2) ===")
print(golden_cross.head(10))

# === 4. Sector Analysis (Module 3) ===
sector_stats = (
    df
    .dropna(subset=['returns'])
    .groupby('sector')['returns']
    .agg(['mean', 'std', 'count'])
    .assign(
        annualized_return=lambda x: x['mean'] * 252,
        annualized_vol=lambda x: x['std'] * np.sqrt(252)
    )
    .sort_values('annualized_return', ascending=False)
)
print("\n=== Sector Analysis (Module 3) ===")
print(sector_stats.round(4))

# === 5. Wide format comparison (Module 6) ===
wide_prices = df.pivot_table(values='price', index='date', columns='ticker')
print("\n=== Wide Format Prices (Module 6) ===")
print(wide_prices.tail())

# === 6. Summary (Module 8 — chaining) ===
print("\n=== Best Stock by Sharpe Ratio (Chained) ===")
summary = (
    df
    .dropna(subset=['returns'])
    .groupby('ticker')
    .agg(
        mean_return=('returns', 'mean'),
        volatility=('returns', 'std'),
        total_days=('returns', 'count')
    )
    .assign(
        ann_return=lambda x: x['mean_return'] * 252,
        ann_vol=lambda x: x['volatility'] * np.sqrt(252),
        sharpe=lambda x: (x['mean_return'] * 252 - 0.04) / (x['volatility'] * np.sqrt(252))
    )
    .sort_values('sharpe', ascending=False)
)
print(summary.round(4))
```

# Quick Reference: Pandas Operations Map

| I need to... | Category / Methods | Common AI Mistakes |
|---|---|---|
| Select columns | `df['col']`, `df[['a','b']]` | Using `df.col` (breaks with spaces/keywords) |
| Filter rows | Boolean masking, `.query()` | Missing parentheses in compound conditions |
| Aggregate by group | `.groupby()` + `.agg()` | Using loops instead of groupby |
| Add group stats to rows | `.transform()` | Using merge when transform is simpler |
| Handle missing data | `.fillna()`, `.dropna()`, `.interpolate()` | Filling with 0 without thinking |
| Combine tables (relate) | `.merge()` with `how` parameter | Not checking for row explosion |
| Stack tables | `pd.concat()` | Confusing concat with merge |
| Reshape long → wide | `.pivot()` or `.pivot_table()` | `pivot` when duplicates exist |
| Reshape wide → long | `.melt()` | Manual column unpacking |
| Time operations | `.resample()`, `.rolling()`, `.pct_change()` | Forgetting to set DatetimeIndex |
