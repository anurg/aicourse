# Module 1: The Figure-Axes Mental Model
*Estimated time: 45 minutes*

## 🧠 CONCEPT: Figure vs Axes

Every Matplotlib confusion traces back to not understanding this hierarchy:

- **Figure:** The entire window or image. Think of it as the canvas.
- **Axes:** A single plot area WITHIN the figure. One figure can have multiple Axes (subplots).
- **Axis:** The actual x-axis or y-axis line with tick marks and labels.

> Figure = the wall, Axes = a picture frame on the wall, Axis = the ruler along the frame edge.

## 🧠 CONCEPT: Two Interfaces — pyplot vs Object-Oriented

**pyplot (plt.plot):** Quick for simple, single plots. Matplotlib tracks the "current" figure behind the scenes.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

**Object-Oriented (fig, ax):** Explicit, reliable, and necessary for anything beyond a single plot. **This is what you should use.**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, np.sin(x), color='steelblue', linewidth=2)
ax.set_title('Sine Wave', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('sin(x)', fontsize=12)
ax.grid(True, alpha=0.3)
plt.show()
```

## 👀 RECOGNIZE: Subplots grid

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, np.sin(x), color='steelblue')
axes[0, 0].set_title('Sine')

axes[0, 1].plot(x, np.cos(x), color='coral')
axes[0, 1].set_title('Cosine')

axes[1, 0].plot(x, np.exp(-x/5) * np.sin(x), color='seagreen')
axes[1, 0].set_title('Damped Sine')

axes[1, 1].plot(x, x**0.5, color='orchid')
axes[1, 1].set_title('Square Root')

fig.suptitle('Four Functions', fontsize=16, fontweight='bold')
fig.tight_layout()
plt.show()
```

### Smell Test #1

AI generates a multi-plot figure mixing `plt.title()` and `ax.set_title()`. **What's the risk?** The `plt` calls modify the "current" axes, which might not be the one you think. In multi-plot figures, labels can appear on the wrong subplot. **If you see `plt` and `ax` mixed, that's a code smell.**

# Module 2: Choosing the Right Chart — The Decision Framework
*Estimated time: 60 minutes*

## 🧠 CONCEPT: Chart selection as analytical thinking

The chart type determines what insight is visible. The wrong chart can hide the most important story in your data.

| Your Question | Chart Type | Example |
|---|---|---|
| How does Y change over time? | Line chart | Stock price over months |
| How do categories compare? | Bar chart (horizontal preferred) | Revenue by department |
| What is the distribution? | Histogram or KDE | Distribution of salaries |
| Is there a relationship between X and Y? | Scatter plot | Ad spend vs revenue |
| What are the proportions? | Stacked bar (NOT pie chart) | Market share by company |
| How does data spread? | Box plot or violin plot | Salary ranges by role |
| How do multiple variables relate? | Heatmap (correlation matrix) | Feature correlations |

> ⚠️ Pie charts are deliberately absent — humans are bad at comparing angles. Use stacked bar charts instead.

## 👀 RECOGNIZE: Distribution visualizations

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate dummy salary data: 1000 employees
rng = np.random.default_rng(42)
salaries = rng.normal(50000, 15000, 1000)
salaries = salaries[salaries > 15000]  # Remove unrealistic values

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

# Histogram: see the shape of distribution
axes[0].hist(salaries, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].set_title('Histogram')
axes[0].set_xlabel('Salary ($)')
axes[0].set_ylabel('Count')

# Box plot: see median, quartiles, outliers
axes[1].boxplot(salaries, vert=True)
axes[1].set_title('Box Plot')
axes[1].set_ylabel('Salary ($)')

# Density (histogram with density=True)
axes[2].hist(salaries, bins=50, density=True, alpha=0.7, color='coral')
axes[2].set_title('Density')
axes[2].set_xlabel('Salary ($)')
axes[2].set_ylabel('Probability Density')

fig.suptitle('Three Ways to View Salary Distribution', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(42)

# --- Scatter: relationship between two continuous variables ---
ad_spend = rng.uniform(1000, 50000, 200)
revenue = ad_spend * 2.5 + rng.normal(0, 10000, 200)  # Noisy linear relationship

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(ad_spend, revenue, alpha=0.5, s=30, color='steelblue')
ax.set_xlabel('Ad Spend ($)')
ax.set_ylabel('Revenue ($)')
ax.set_title('Ad Spend vs Revenue')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# Module 3: Customization That Matters — Labels, Colors, and Clarity
*Estimated time: 45 minutes*

## 🧠 CONCEPT: Every visual element should earn its place

Every chart must have these 5 non-negotiables:
1. **Title** — specific, not vague
2. **Axis labels with units** — "Revenue ($K)" not just "Revenue"
3. **Readable tick marks**
4. **Legend (if multiple series)**
5. **Appropriate figure size** — `figsize=(10, 6)` is a good starting point

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate dummy monthly revenue for two product lines
rng = np.random.default_rng(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_a = rng.integers(40, 80, 12)  # Revenue in $K
product_b = rng.integers(30, 70, 12)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(months, product_a, label='Product A', linewidth=2, marker='o')
ax.plot(months, product_b, label='Product B', linewidth=2, marker='s')
ax.set_title('Monthly Revenue by Product Line, 2024',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Revenue ($K)', fontsize=12)
ax.legend(loc='upper left', frameon=True)
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## 🧠 CONCEPT: Color as information channel

- **Sequential colormaps** (low→high): `'viridis'`, `'plasma'`, `'cividis'` — colorblind-safe
- **Diverging colormaps** (meaningful center): `'RdBu'`, `'coolwarm'` — two colors from neutral middle
- **Categorical colors** (distinct groups): `'tab10'` — max 7-8 colors

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Heatmap with diverging colormap (correlation matrix) ---
rng = np.random.default_rng(42)
base = rng.normal(0, 1, 200)
data = np.column_stack([
    base + rng.normal(0, 0.3, 200),
    -base + rng.normal(0, 0.5, 200),
    rng.normal(0, 1, 200),
    base * 0.5 + rng.normal(0, 0.7, 200),
    rng.normal(0, 1, 200),
])
labels = ['Revenue', 'Cost', 'Headcount', 'Profit', 'Market Cap']
corr = np.corrcoef(data.T)

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(corr, cmap='RdBu', vmin=-1, vmax=1)
fig.colorbar(im, ax=ax, label='Correlation')
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels)

# Add values as text
for i in range(len(labels)):
    for j in range(len(labels)):
        ax.text(j, i, f'{corr[i, j]:.2f}', ha='center', va='center', fontsize=9)

ax.set_title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Color encoding meaning: red for loss, green for gain ---
rng = np.random.default_rng(42)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
returns = rng.normal(0, 2, 5)  # Daily returns in %
colors = ['#d32f2f' if v < 0 else '#388e3c' for v in returns]

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(days, returns, color=colors, edgecolor='white', width=0.6)
ax.axhline(y=0, color='black', linewidth=0.8)
ax.set_ylabel('Daily Return (%)')
ax.set_title('Daily Portfolio Returns', fontsize=14, fontweight='bold')
ax.grid(True, axis='y', alpha=0.3)
fig.tight_layout()
plt.show()
```

# Module 4: Statistical Visualization — Seeing Patterns
*Estimated time: 60 minutes*

## 🧠 CONCEPT: Visualization as analysis

**Anscombe's Quartet** is the classic demonstration: four datasets with identical summary statistics but completely different visual patterns. You can't understand data from numbers alone; you must look at it.

## 🧠 CONCEPT: Understanding data shape — Bins Matter

```python
import matplotlib.pyplot as plt
import numpy as np

# Bimodal salary data: two departments with different pay scales
rng = np.random.default_rng(42)
salaries = np.concatenate([
    rng.normal(45000, 8000, 500),   # Department A
    rng.normal(75000, 10000, 300),  # Department B
])

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

axes[0].hist(salaries, bins=10, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].set_title('10 Bins — Hides the two groups!')
axes[0].set_xlabel('Salary ($)')

axes[1].hist(salaries, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[1].set_title('30 Bins — Two peaks visible')
axes[1].set_xlabel('Salary ($)')

axes[2].hist(salaries, bins=80, edgecolor='black', alpha=0.7, color='steelblue')
axes[2].set_title('80 Bins — Too noisy')
axes[2].set_xlabel('Salary ($)')

fig.suptitle('Same Data, Different Bin Counts', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

## 🧠 CONCEPT: Visual comparison techniques

```python
import matplotlib.pyplot as plt
import numpy as np

# Two groups of test scores
rng = np.random.default_rng(42)
group_a = rng.normal(72, 12, 500)  # Mean 72, std 12
group_b = rng.normal(78, 10, 500)  # Mean 78, std 10 (slightly better)

# --- Overlapping histograms ---
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(group_a, bins=30, alpha=0.5, label='Group A (old method)', color='steelblue')
ax.hist(group_b, bins=30, alpha=0.5, label='Group B (new method)', color='coral')
ax.legend(fontsize=11)
ax.set_xlabel('Test Score')
ax.set_ylabel('Count')
ax.set_title('Score Distribution: Old Method vs New Method', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## 🧠 CONCEPT: Seeing relationships — Scatter with trend line

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate correlated data: study hours vs exam score
rng = np.random.default_rng(42)
study_hours = rng.uniform(1, 10, 150)
exam_score = study_hours * 7 + 30 + rng.normal(0, 8, 150)  # Noisy linear
exam_score = np.clip(exam_score, 0, 100)

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(study_hours, exam_score, alpha=0.5, s=30, color='steelblue')

# Add trend line
z = np.polyfit(study_hours, exam_score, 1)
p = np.poly1d(z)
x_line = np.linspace(1, 10, 100)
r = np.corrcoef(study_hours, exam_score)[0, 1]
ax.plot(x_line, p(x_line), 'r--', linewidth=2,
        label=f'Trend line (r = {r:.2f})')

ax.set_xlabel('Study Hours per Day')
ax.set_ylabel('Exam Score')
ax.set_title('Study Hours vs Exam Performance', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

# Module 5: Time Series Visualization
*Estimated time: 45 minutes*

## 🧠 CONCEPT: Time has unique visual challenges

Time series plots are the bread and butter of financial analysis. This connects directly to your investing work.

## 👀 RECOGNIZE: Line plot with dates

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd

# Simulate 1 year of daily stock prices
rng = np.random.default_rng(42)
dates = pd.date_range('2024-01-01', periods=252, freq='B')
prices = 150 + np.cumsum(rng.normal(0.05, 2, 252))

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(dates, prices, linewidth=1.2, color='steelblue')

# Format dates nicely on x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')

ax.set_title('Stock Price History (Simulated)', fontsize=14, fontweight='bold')
ax.set_ylabel('Price ($)')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## 🧠 CONCEPT: Layering information — Moving Averages

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd

# Simulate stock data with enough history for 50-day MA
rng = np.random.default_rng(42)
dates = pd.date_range('2023-06-01', periods=400, freq='B')
prices = 100 + np.cumsum(rng.normal(0.03, 1.5, 400))

# Compute moving averages using Pandas
df = pd.DataFrame({'price': prices}, index=dates)
df['MA_20'] = df['price'].rolling(window=20).mean()
df['MA_50'] = df['price'].rolling(window=50).mean()

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(df.index, df['price'], label='Price', alpha=0.7, linewidth=1, color='gray')
ax.plot(df.index, df['MA_20'], label='20-day MA', linewidth=1.8, color='steelblue')
ax.plot(df.index, df['MA_50'], label='50-day MA', linewidth=1.8, color='coral')

# Shade green when 20MA > 50MA (bullish), red when below (bearish)
ax.fill_between(df.index, df['MA_20'], df['MA_50'],
                where=df['MA_20'] > df['MA_50'],
                alpha=0.15, color='green', label='Bullish')
ax.fill_between(df.index, df['MA_20'], df['MA_50'],
                where=df['MA_20'] <= df['MA_50'],
                alpha=0.15, color='red', label='Bearish')

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')

ax.legend(loc='upper left')
ax.set_title('Price with Moving Average Crossover', fontsize=14, fontweight='bold')
ax.set_ylabel('Price ($)')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## 🧠 CONCEPT: Dual Y-Axes — Use With Caution

Dual-axis charts can imply a relationship that doesn't exist. Use them only when the two variables are genuinely related (like price and volume).

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulate price + volume data
rng = np.random.default_rng(42)
dates = pd.date_range('2024-01-01', periods=120, freq='B')
prices = 100 + np.cumsum(rng.normal(0.05, 1.2, 120))
volume = rng.integers(500000, 2000000, 120)
volume[50:55] = rng.integers(3000000, 5000000, 5)  # Volume spike

fig, ax1 = plt.subplots(figsize=(12, 5))

# First y-axis: Price
color1 = 'steelblue'
ax1.plot(dates, prices, color=color1, linewidth=1.5, label='Price')
ax1.set_ylabel('Price ($)', color=color1, fontsize=12)
ax1.tick_params(axis='y', labelcolor=color1)

# Second y-axis: Volume
ax2 = ax1.twinx()
color2 = 'coral'
ax2.bar(dates, volume, color=color2, alpha=0.3, width=1.5, label='Volume')
ax2.set_ylabel('Volume', color=color2, fontsize=12)
ax2.tick_params(axis='y', labelcolor=color2)

ax1.set_title('Price and Volume (Dual Axis)', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

# Module 6: Annotations, Highlights, and Storytelling
*Estimated time: 30 minutes*

## 🧠 CONCEPT: Charts should answer questions, not just show data

Annotations, highlights, and reference lines transform a chart from "here is data" to "here is what you should notice."

## 👀 RECOGNIZE: Common annotation patterns

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd

# Simulate stock data with a clear event
rng = np.random.default_rng(42)
dates = pd.date_range('2024-01-01', periods=252, freq='B')
prices = 100 + np.cumsum(rng.normal(0.02, 1.0, 252))

# Simulate earnings beat at day 80 (price jumps)
prices[80:] += 15
# Simulate a dip at day 180
prices[180:200] -= np.linspace(0, 8, 20)
prices[200:] -= 8

target_price = 130
earnings_date = dates[80]
peak_idx = np.argmax(prices)
peak_price = prices[peak_idx]
dip_start = dates[180]
dip_end = dates[200]

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(dates, prices, linewidth=1.2, color='steelblue')

# Reference line (target price)
ax.axhline(y=target_price, color='red', linestyle='--',
           linewidth=1, label='Target Price')

# Vertical line to mark earnings date
ax.axvline(x=earnings_date, color='gray', linestyle=':', alpha=0.7)

# Annotate the earnings event with arrow
ax.annotate('Earnings Beat!',
            xy=(earnings_date, prices[80]),
            xytext=(dates[100], prices[80] + 15),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=11, fontweight='bold')

# Shade a region (correction period)
ax.axvspan(dip_start, dip_end, alpha=0.1, color='red', label='Correction')

# Text box on the plot
ax.text(0.02, 0.95, 'Simulated Stock — 2024',
        transform=ax.transAxes, fontsize=11,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')

ax.legend(loc='lower right')
ax.set_title('Annotated Stock Chart', fontsize=14, fontweight='bold')
ax.set_ylabel('Price ($)')
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

# Module 7: Common Pitfalls and How to Spot Them
*Estimated time: 45 minutes*

## 🧠 CONCEPT: Visual literacy as a critical skill

Charts can mislead, even unintentionally. AI will sometimes produce charts with these problems.

### Pitfall 1: Truncated Y-Axis

```python
import matplotlib.pyplot as plt
import numpy as np

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
satisfaction = [96.2, 97.1, 96.8, 97.5]  # Tiny differences

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# LEFT: Misleading — truncated y-axis makes differences look huge
axes[0].bar(quarters, satisfaction, color='steelblue', edgecolor='white')
axes[0].set_ylim(95.5, 98)  # Truncated!
axes[0].set_title('MISLEADING: Truncated Y-Axis', fontsize=13, color='red')
axes[0].set_ylabel('Satisfaction Score')

# RIGHT: Honest — starts from 0, shows true proportion
axes[1].bar(quarters, satisfaction, color='steelblue', edgecolor='white')
axes[1].set_ylim(0, 100)
axes[1].set_title('HONEST: Full Y-Axis (0-100)', fontsize=13, color='green')
axes[1].set_ylabel('Satisfaction Score')

fig.suptitle('Same Data, Different Story', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

### Pitfall 2: Cherry-Picked Time Range

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Stock that crashed then partially recovered
rng = np.random.default_rng(42)
dates = pd.date_range('2022-01-01', periods=750, freq='B')

prices = 200 + np.cumsum(rng.normal(0.01, 1.5, 750))
prices[300:400] -= np.linspace(0, 60, 100)  # Crash
prices[400:] -= 60
prices[500:] += np.cumsum(rng.normal(0.15, 0.8, 250))  # Recovery

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# LEFT: Cherry-picked — only shows the recovery
axes[0].plot(dates[550:], prices[550:], color='green', linewidth=1.5)
axes[0].set_title('CHERRY-PICKED: Only Recent Recovery', fontsize=12, color='red')
axes[0].set_ylabel('Price ($)')
axes[0].tick_params(axis='x', rotation=45)

# RIGHT: Full picture
axes[1].plot(dates, prices, color='steelblue', linewidth=1.2)
axes[1].axvspan(dates[550], dates[-1], alpha=0.1, color='green', label='Cherry-picked range')
axes[1].set_title('HONEST: Full History', fontsize=12, color='green')
axes[1].set_ylabel('Price ($)')
axes[1].legend()
axes[1].tick_params(axis='x', rotation=45)

fig.suptitle('Time Range Changes the Story', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

### Pitfall 3: Overplotting

```python
import matplotlib.pyplot as plt
import numpy as np

# 50,000 clustered points
rng = np.random.default_rng(42)
x = np.concatenate([rng.normal(2, 1, 25000), rng.normal(6, 1.5, 25000)])
y = np.concatenate([rng.normal(3, 1, 25000), rng.normal(7, 1, 25000)])

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# LEFT: Overplotted blob
axes[0].scatter(x, y, s=5, color='steelblue')
axes[0].set_title('PROBLEM: Overplotted (50K points)', color='red')

# MIDDLE: Alpha transparency reveals density
axes[1].scatter(x, y, alpha=0.02, s=5, color='steelblue')
axes[1].set_title('FIX 1: Alpha Transparency', color='green')

# RIGHT: Hexbin shows density clearly
hb = axes[2].hexbin(x, y, gridsize=30, cmap='viridis')
axes[2].set_title('FIX 2: Hexbin Density', color='green')
fig.colorbar(hb, ax=axes[2], label='Count')

fig.suptitle('Three Ways to Handle 50K Points', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

### Pitfall 4: Too Many Colors/Lines

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulate 10 stocks
rng = np.random.default_rng(42)
dates = pd.date_range('2024-01-01', periods=120, freq='B')
stocks = {}
for name in ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'META',
             'TSLA', 'NFLX', 'NVDA', 'AMD', 'INTC']:
    stocks[name] = 100 + np.cumsum(rng.normal(0.05, 1.5, 120))

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# LEFT: Unreadable spaghetti
for name, prices in stocks.items():
    axes[0].plot(dates, prices, linewidth=1.5, label=name)
axes[0].legend(fontsize=7)
axes[0].set_title('PROBLEM: 10 Lines = Spaghetti', color='red')
axes[0].tick_params(axis='x', rotation=45)

# RIGHT: Highlight one, fade the rest
for name, prices in stocks.items():
    if name == 'NVDA':
        axes[1].plot(dates, prices, linewidth=2.5, color='red', label='NVDA')
    else:
        axes[1].plot(dates, prices, linewidth=0.5, color='gray', alpha=0.4)
axes[1].legend(fontsize=11)
axes[1].set_title('FIX: Highlight Key Series', color='green')
axes[1].tick_params(axis='x', rotation=45)

fig.suptitle('Managing Visual Complexity', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

### Pitfall 5: Missing Context / Benchmark

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Portfolio vs benchmark
rng = np.random.default_rng(42)
dates = pd.date_range('2024-01-01', periods=252, freq='B')
portfolio = 100 * np.cumprod(1 + rng.normal(0.0003, 0.01, 252))  # Your portfolio
benchmark = 100 * np.cumprod(1 + rng.normal(0.0004, 0.008, 252)) # S&P 500

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# LEFT: No context — looks great in isolation!
axes[0].plot(dates, portfolio, color='steelblue', linewidth=1.5)
axes[0].set_title('MISLEADING: Portfolio Going Up!', color='red')
axes[0].set_ylabel('Value ($)')
axes[0].tick_params(axis='x', rotation=45)

# RIGHT: With benchmark — tells the true story
axes[1].plot(dates, portfolio, color='steelblue', linewidth=1.5, label='My Portfolio')
axes[1].plot(dates, benchmark, color='coral', linewidth=1.5,
             linestyle='--', label='S&P 500 Benchmark')
axes[1].set_title('HONEST: Portfolio vs Benchmark', color='green')
axes[1].set_ylabel('Value ($)')
axes[1].legend(fontsize=11)
axes[1].tick_params(axis='x', rotation=45)

fig.suptitle('Context Changes Everything', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

### Key pattern: Pass `ax=ax` to combine Pandas convenience with Matplotlib customization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Combining Pandas .plot() with Matplotlib customization
rng = np.random.default_rng(42)
months = pd.date_range('2024-01-01', periods=12, freq='ME')
revenue = pd.Series(rng.integers(100, 250, 12), index=months)

fig, ax = plt.subplots(figsize=(10, 5))
# Convert DatetimeIndex to month strings to avoid freq issues with .plot.bar()
revenue_plot = revenue.copy()
revenue_plot.index = [d.strftime('%b') for d in revenue.index]
revenue_plot.plot.bar(ax=ax, color='steelblue', edgecolor='white')  # Pass ax!
ax.set_title('Monthly Revenue, 2024', fontsize=14, fontweight='bold')
ax.set_ylabel('Revenue ($K)')
ax.set_xticklabels(revenue_plot.index, rotation=45, ha='right')
ax.axhline(y=revenue.mean(), color='red', linestyle='--', label=f'Avg: {revenue.mean():.0f}')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
fig.tight_layout()
plt.show()
```

# Capstone Project: Investment Dashboard
*Estimated time: 3-4 hours*

Build a comprehensive visual analysis of a stock portfolio. Fill in each panel using techniques from the course.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd

# ---- Generate comprehensive dummy portfolio data ----
rng = np.random.default_rng(42)
dates = pd.date_range('2023-01-01', periods=500, freq='B')
price = 100 + np.cumsum(rng.normal(0.04, 1.5, 500))

df = pd.DataFrame({'price': price}, index=dates)
df['MA_20'] = df['price'].rolling(20).mean()
df['MA_50'] = df['price'].rolling(50).mean()
df['returns'] = df['price'].pct_change()
df['volatility_30d'] = df['returns'].rolling(30).std() * np.sqrt(252)
df['cumulative_max'] = df['price'].cummax()
df['drawdown'] = (df['price'] - df['cumulative_max']) / df['cumulative_max'] * 100
df['volume'] = rng.integers(500000, 2000000, 500)

# ---- Create the 2x3 Dashboard ----
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Panel 1: Price with Moving Averages
# WHY line chart: shows trend over time
axes[0, 0].plot(df.index, df['price'], alpha=0.5, linewidth=0.8, color='gray', label='Price')
axes[0, 0].plot(df.index, df['MA_20'], linewidth=1.5, color='steelblue', label='20-day MA')
axes[0, 0].plot(df.index, df['MA_50'], linewidth=1.5, color='coral', label='50-day MA')
axes[0, 0].set_title('Price & Moving Averages')
axes[0, 0].set_ylabel('Price ($)')
axes[0, 0].legend(fontsize=8)
axes[0, 0].grid(True, alpha=0.3)

# Panel 2: Daily Returns Distribution (Histogram)
# WHY histogram: reveals distribution shape
clean_returns = df['returns'].dropna()
axes[0, 1].hist(clean_returns, bins=50, edgecolor='black', alpha=0.7, color='steelblue')
axes[0, 1].axvline(x=clean_returns.mean(), color='red', linestyle='--', label=f'Mean: {clean_returns.mean():.4f}')
axes[0, 1].set_title('Daily Returns Distribution')
axes[0, 1].set_xlabel('Return')
axes[0, 1].set_ylabel('Count')
axes[0, 1].legend(fontsize=8)

# Panel 3: Rolling 30-Day Volatility
# WHY line chart: shows how risk changes over time
axes[0, 2].plot(df.index, df['volatility_30d'], linewidth=1.2, color='coral')
max_vol_idx = df['volatility_30d'].idxmax()
axes[0, 2].annotate('Peak Vol', xy=(max_vol_idx, df.loc[max_vol_idx, 'volatility_30d']),
                     fontsize=9, fontweight='bold',
                     arrowprops=dict(arrowstyle='->', color='black'),
                     xytext=(max_vol_idx + pd.Timedelta(days=30),
                             df['volatility_30d'].max() * 0.85))
axes[0, 2].set_title('Rolling 30-Day Volatility (Annualized)')
axes[0, 2].set_ylabel('Volatility')
axes[0, 2].grid(True, alpha=0.3)

# Panel 4: Drawdown Chart
# WHY area chart: shows depth and duration of losses
axes[1, 0].fill_between(df.index, df['drawdown'], 0, alpha=0.5, color='red')
axes[1, 0].plot(df.index, df['drawdown'], linewidth=0.5, color='darkred')
worst_dd_idx = df['drawdown'].idxmin()
axes[1, 0].annotate(f'Max: {df["drawdown"].min():.1f}%',
                     xy=(worst_dd_idx, df.loc[worst_dd_idx, 'drawdown']),
                     fontsize=9, fontweight='bold')
axes[1, 0].set_title('Drawdown')
axes[1, 0].set_ylabel('Drawdown (%)')
axes[1, 0].grid(True, alpha=0.3)

# Panel 5: Volume Bars
# WHY bar chart: shows trading activity over time
axes[1, 1].bar(df.index, df['volume'], width=1.5, color='steelblue', alpha=0.5)
axes[1, 1].set_title('Trading Volume')
axes[1, 1].set_ylabel('Volume')
axes[1, 1].grid(True, alpha=0.3)

# Panel 6: Monthly Returns Bar
# WHY bar chart: compare performance across months
monthly_returns = df['returns'].resample('ME').sum() * 100
colors_monthly = ['#388e3c' if v >= 0 else '#d32f2f' for v in monthly_returns]
axes[1, 2].bar(range(len(monthly_returns)), monthly_returns, color=colors_monthly, edgecolor='white')
axes[1, 2].axhline(y=0, color='black', linewidth=0.5)
axes[1, 2].set_title('Monthly Returns (%)')
axes[1, 2].set_ylabel('Return (%)')
axes[1, 2].set_xlabel('Month')
axes[1, 2].grid(True, axis='y', alpha=0.3)

fig.suptitle('Portfolio Dashboard — 2023-2024',
             fontsize=16, fontweight='bold', y=1.01)
fig.tight_layout()
fig.savefig('dashboard.png', dpi=150, bbox_inches='tight', facecolor='white')
fig.savefig('dashboard.pdf', bbox_inches='tight')
plt.show()
print("Dashboard saved as dashboard.png and dashboard.pdf")
```
