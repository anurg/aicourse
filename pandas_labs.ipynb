```python
#Pandas: For Data Analysis and Data Manipulations

#here we have 2 categories
#1.series(1-D data) :nothing but a column
#2.DataFrames(2-D Tabular data): Collection of Series(columns)

#creating a series
#creating a series from a list
import pandas as pd
x=[1,2,3,4,5]
z=pd.Series(x)
print(z)

# creating our own indexes
z=pd.Series(x,index=["a","b","c","d","e"])
print(z)
```

```python
# creating a Series from dictionary
d ={"name":"Anurag","age":51,"city":"Delhi"}
z=pd.Series(d)
print(z)

# accessing element from Series
print(z["city"])
```

```python
# Creating DataFrames

#i) Creating Dataframes from a Dictionary
import pandas as pd
#ex:

dict1={'Inter-I':[90,80,70,40,50],
       'Inter-II':[50,40,55,45,40]}
df=pd.DataFrame(dict1)
print(df)
df=pd.DataFrame(dict1,index=["Jan","Feb","Mar","Apr","May"])
print(df)
# Retrieving particular month data in a row
print(df.loc["Jan"])
#4) Indexing and slicing

#a)printing a specific column(series)
print(df["Inter-I"])
print(df["Inter-I"]["Jan"])
#b)Printing a Dataframe
print(df[["Inter-I"]])
print(df[["Inter-I","Inter-II"]])
#d)Slicing a Dataframe
#  Retrieving specific rows
print(df[0:3])
print(df.loc[["Jan","Mar"]])
print(df.iloc[[0,3]])
print(df.iloc[0:3,0:1])
print(df.iloc[:,0:1])
```

```python
import pandas as pd
data={
      "Sony":[250,350,200,350,300],
      "LG":[300,180,260,180,150],
      "Lenovo":[300,180,260,200,150]
      }
df= pd.DataFrame(data,index=["Jan","Feb","Mar","Apr","May"])
print(df)
print(df.loc["Jan"]) #Retrieving particular month data(particular row)
print(df.loc[["Jan","May"]])
print(df["LG"])   #here we see no colname
print(df[["LG"]]) #b)Printing a Dataframe
print(df[['LG','Lenovo']])
print(df[0:3]) #  Retrieving specific rows

#e)More data selection operations
#Using loc and iloc, you can select certain rows in a data set.
#loc uses string indices
#iloc uses integers.

print(df.loc[["Mar","May"]])
print(df.iloc[3]) #f) iloc
print(df.iloc[1:3]) # 1st and 2nd row, all columns
print(df.iloc[:,:]) # all rows and all columns
print(df.iloc[:,1:3]) # all rows and 2 columns
print(df.iloc[0:2,1:3])
print(df.iloc[0:2,1:3])
print(df.iloc[0:3,0:3:2]) # skip columns by 2
print(df.iloc[0:3:2,0:3:2]) # skip row and columns by 2
#To eaxtract--->250 from the 1st row
print(df.iloc[0:1,0:1])
```

```python
#h) sorting data
#   df.sort_values()
import pandas as pd
data={
      "Sony":[250,350,200,350,300],
      "LG":[300,180,260,180,150],
      "Lenovo":[300,180,260,200,150]
      }
df = pd.DataFrame(data,index=["April","May","June","July","August"])
print(df)
print(df.sort_values(["LG","Lenovo"]))
print(df.sort_values(["LG","Lenovo"],ascending=True))
#i) Renaming a dataframe column
#   Renaming Sony to Samsung
df.columns=["Samsung","LG","Leno"]
print(df)
print(df.drop("Samsung",axis=1)) #drop a column
print(df)
print(df.drop("May")) #drop a row
#Adding new columns
df["Oppo"] = df["Leno"]+10
df["Mi"] = df.eval("Samsung + LG")
df["Realme"] = pd.Series([350,170,250,200,370],index=["April","May","June","July","August"])
print(df)
```
